```python
import ast
```

```python
def convert_and_append_names(obj):
    try:
        literal_object = ast.literal_eval(obj)
        if isinstance(literal_object, list):
            names = [item.get('name', '') for item in literal_object]
            return names
    except (SyntaxError, ValueError):
        return []
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert_and_append_names)
movies_df['keywords'] = movies_df['keywords'].apply(convert_and_append_names)
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
        return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert3)
```

```python
def director(obj):
    try:
        literal_object = ast.literal_eval(obj)
        if isinstance(literal_object, list):
            names = [item.get('name') for item in literal_object if item.get('job') == 'Director']
            return names
    except (SyntaxError, ValueError):
        return []
```

```python
movies_df['crew'] = movies_df['crew'].apply(director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```
